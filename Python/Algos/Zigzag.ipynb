```python
def convert(s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        if numRows == 1 or numRows >= len(s):
            return s

        L = [''] * numRows
        print(L)
        index, step = 0, 1

        for x in s:
            print("--",x,"--")
            L[index] += x
            print(L[index])
            if index == 0:
                step = 1
            elif index == numRows -1:
                step = -1
            index += step

        return ''.join(L)
        
convert("PAYPALISHIRING",3)
convert("idouvypdupfyejxgqtuorkyqyvnpmutwxhqufgazxfzbqzigseulrubpqreelyakhqhpyhqvoqzepjl",2)
```
